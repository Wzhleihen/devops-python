```python
def triangular_print(n):
    for i in range(1, n + 1):
        for j in range(n,0, -1):
            print(j, end='')
        print()
triangular_print(5)
```

```python
def triangular_print_v1(n):
    for i in range(1, n + 1):
        for j in range(n,0, -1):
            # if j <= i:
            #     print(j, end=' ')
            # else:
            #     print(len(str(j)) * ' ', end=' ')
            print(len(str(j)) * ' ' if j > i else j , end=' ')
        print()
```

```python
def triangular_print_v2(n):
    for i in range(n ,0, -1):
        for j in range(n,0, -1):
            # if j <= i:
            #     print(j, end=' ')
            # else:
            #     print(len(str(j)) * ' ', end=' ')
            print(len(str(j)) * ' ' if j > i else j , end=' ')
        print()
triangular_print_v2(12)
```

```python
triangular_print_v1(12)
triangular_print_v2(12)
```

```python
def triangular_print_v3(n):
    tail = ' '.join(map(str, range(n, 0, -1)))
    width = len(tail)
    for i in range(width, -1, -1):
        if tail[i] == ' ':
            print("{:>{}}".format(tail[i+1:], width))
    print(tail)
```

```python
triangular_print_v3(12)
```
