````````python
nums = [9, 8, 1, 2, 3, 4, 5, 6, 7]
length = len(nums)
count = 0
count_swap = 0


for i in range(length -1):
    swapped = False  # 优化点
    for j in range(length-1-i):
        if nums[j] > nums[j+1]:
            nums[j] , nums[j+1] = nums[j+1], nums[j]
            swapped = True  # # 有一次交换就要标记为True
            count_swap += 1  #交换计数
        count += 1  # 比较计数
    # 当没有发生交换时，说明已经排序完成
    if not swapped:  
        break
    print(f'第{i+1}次', nums)
print(count, count_swap)
````````

````````python
def fib_v1(n):
    a = b = 1
    for i in range(n-2):
        a, b = b, a + b
    return b

fib_v1(101)
````````

````````python
%timeit fib_v1(101)
````````

````````python
def fib_v2(n):
    if n == 1 or n == 2:
        return 1
    return fib_v2(n-1) + fib_v2(n-2)
````````

````````python
%timeit fib_v2(101)
````````

````````python
def get_values_v1(x, y, *args):
    return min(x, y, *args), max(x, y, *args)
````````

````````python
def get_values_v2(x, y, *args):
    t = (x, y, *args)
    return min(t), max(t)
````````

````````python
def get_values_v3(x, y, *args):
    min, max = (x , y) if x < y else (y,x)
    for i in args:
        if i > max:
            max = i
        if i < min:
            min = i
        return min, max 
````````
