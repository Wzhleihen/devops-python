# 上下文管理
context 目前总结的是针对文件对象的，不适用其他支持上下文的对象

with 文件对象 as f1：  # f1文件对象的别名
    pass

with 离开的时候，文件对象会调用close方法，不管with中是否正常还是异常退出（不正常执行了）

```python
f = open('test', encoding='utf-8')
print(f.closed)
with f:
    print(f.closed)
    I\O 
    print('-----------')
```

```python
f.closed
```

```python
# 文件对象比较
# 每次打开都的到不一样的对象，地址不一样，但是内容是一样的，比较为false

import sys

f2 = open('test')
with f2 as f1:
    print(f1)
    print(f2)
    print(f1 == f2)
    print(f1 is f2)

print(f2.closed)
```

```python
f3 = open('test')
f4 = open('test')
# 两次打开得到两次文件描述符，文件描述符不一样，但是文件内容是一样的，比较为false

print(f3 == f4)  # 文件对象不可以直接比较内容，is是比较id
print(id(f3))
print(id(f4))

# 返回该文件对象在操作系统中的文件描述符编号
print(f3.fileno(), f4.fileno())


f3.close()
f4.close()
print(f3.closed)
print(f4.closed)
```

# 路径操作
## os.path模块

```python
#第一种导入方式
import os
os.path



```

```python
# 第二种导入方式
from os import path
path

# 标识符 path 识别对应的模块
```

```python
# 常见的使用场景
path.join('/etc', 'sysconfig', 'network-scripts', 'ifcfg-eth0')
# 在windows下，会自动转换成\\
```

```python
# 获取目录名和基名
path.dirname('/etc/sysconfig/network-scripts/ifcfg-eth0')  # 获取目录名
path.basename('/etc/sysconfig/network-scripts/ifcfg-eth0')
```

```python
# 分割路径
path.split('/etc/sysconfig/network-scripts/ifcfg-eth0')  # 将路径分割成目录和文件名，获得元组
```

```python
# 在windows下，将路径分割成盘符和路径，获得元组
path.splitdrive('d:/etc/sysconfig/') 
```

```python
# 获取扩展名
path.splitext('/etc/sysconfig/network-scripts/eth0.yaml')
```

```python
# 判断是否是目录
path.isdir('/etc/sysconfig/network-scripts')  
```

```python
# 获取绝对路径
path.abspath('sysconfig')  
```
