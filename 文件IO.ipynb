```python
f = open("test") # file对象
# windows <_io.TextIOWrapper name='test' mode='r' encoding='cp936'>
# linux <_io.TextIOWrapper name='test' mode='r' encoding='UTF-8'>
print(f.read()) # 读取文件
f.close() # 关闭文件,flush 2 归还文件描述符

# open操作成功打开后，会占用当前应用程序进程一个描述符
''' 
每个进程都有一个文件描述符表（file descriptor table）
open() 成功后会向该表中增加一项
对应的整数 FD（文件描述符）由操作系统分配，例如 3、4、5…
'''
```
