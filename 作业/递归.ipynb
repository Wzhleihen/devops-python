猴子第一天摘下若干个桃子，当即吃了一半，还不过瘾，又多吃了一个。
第二天早上又将剩下的桃子吃掉一半，又多吃了一个。以后每天早上都吃了前一天剩下的一半零一个。
到第10天早上想吃时，只剩下一个桃子了。求第一天共摘多少个桃子

```python
# 分析：
# 1. 逆向思维，从第10天开始计算  day10 = 1
# 2. 第10天早上剩下一个桃子，所以第9天早上有2个桃子  day9 = (1+1)*2
# 3. 第9天早上有4个桃子，所以第8天早上有(4+1)*2=10个桃子 day8 = (4+1)*2
# 若第 n 天早上剩 f(n) 个桃子，

# 则前一天（第 n-1 天）的桃子数量满足：f(n-1) = (f(n) + 1) * 2

```

```python
#  # 第 n 天早上剩 f(n) 个桃子，所以第 n-1 天早上有 (f(n) + 1) * 2 个桃子
x = 1
for i in range(9):
    x = 2 * (x + 1)
print(x)
```

```python

# 递归函数实现
def monkey_peach(day):
    # 边界条件：第10天早上剩1个
    if day == 10:
        return 1
     # 递归关系：前一天的桃子 = （后一天桃子 + 1）* 2
    else:
        return (monkey_peach(day + 1) + 1) * 2

print(monkey_peach(9))
```

 n 的阶乘
# 分析：
# 1. 递归关系：n! = n * (n-1)!
# 2. 边界条件：0! = 1


```python

# 编写一个递归函数来计算 n 的阶乘
def factorial_v1(n):
    # 边界条件：0的阶乘为1
    if n == 0:
        return 1
    # 递归关系：n的阶乘 = n * (n-1)的阶乘
    if n < 3:
        return n
    else:
        return n * factorial_v1(n - 1)

# 循环实现
def factorial_v2(n):
    s = 1
    for i in range(2, n+1):
        s *= i
    return s

# 第二种循环实现
def factorial_v3(n):
    s = 1
    for i in range(n, 0, -1):
        s *= i
    return s


# 递归循环
def factorial_v4(n, s=1): # s=1是默认参数，默认值为1
    if n == 1:
        return s
    return factorial_v4(n-1, s * n) # f(2-1, 1*2)
    
    
print(factorial_v3(5))
print(factorial_v2(4))
print(factorial_v1(5))
```
