```python
def counter(base):
    def inc(step=1):
        nonlocal base
        base += step
        return base
    return inc
""" 
因为如果不加 nonlocal：
Python 会把 base 当做 inc() 内部的局部变量
增加 base += step 时会报错（未绑定引用）
nonlocal base 的作用是：
告诉 Python：base 不是本地变量，请使用外层函数中的变量。


"""
f = counter(10)
print(f())
print(f())

''' 
闭包：内部函数和外部函数变量环境 运行结束，还有未定义的变量存在内存中没有释放，这些变量被cpu记住内存地址，可以被再次调用

闭包（Closure）概念总结：

1. 闭包 = 内部函数 + 外部函数的变量环境（未释放的局部变量）
2. 即使外部函数 counter() 已经执行结束，
   它的变量 base 仍然被内部函数 inc() 捕获并保存在内存中。
3. 每次调用 f()（也就是 inc）时，
   都可以继续访问和修改这个外层变量 base。
counter(10)
   ↓
返回 inc 函数 + 保存 base=10
   ↓
f() → base=11
f() → base=12

'''
```

```python
def counter(base):
    def inc(step=1): # 有没有闭包？
        nonlocal base # 形参base也是外部函数counter的local变量
        base += step
        return base
    return inc
# c1 = counter(5)

# print(c1())
# print(c1())
# print(c1())
f1 = counter(5)
f2 = counter(5)
print(f1 == f2) # 相等吗？
print(f1 is f2)  # False

# 原因
''' 
每次 counter(5) 都会创建 新的闭包实例
关键点：闭包对象之间永远不可能相等
f1 的 base 是独立的闭包变量环境
f2 的 base 是另一份独立的闭包变量环境

想更清楚？看它们抓住的自由变量
你会看到两个不同的 cell object 地址。
'''
print(f1.__closure__)
print(f2.__closure__)
```
