```python
import inspect
from functools import wraps

def check(fn):
    @wraps(fn)
    def wrapper(*args, **kwargs):
        params = inspect.signature(fn).parameters
        
        # values = tuple(params.values())
        # for i,v in enumerate(args): # 迭代谁？
        for v,p in zip(args, params.values()):
            if p.annotation != p.empty and not isinstance(v, p.annotation):
                print(f"{p.name} = {v} .ag is not OK")
                # raise TypeError("{p.name} = {v} is not OK")
        
        for k,v in kwargs.items():
            if params[k].annotation != inspect._empty and  not isinstance(v, params[k].annotation):
                print(f"{params[k].name} = {v} .kw is not OK")

        ret = fn(*args, **kwargs)
        return ret
    return wrapper

@check
def add(x:int, y:int = 7):
    print('------')
    return x + y


# add('4','5')
# add('4',y='5')
add(x='4', y='5')

```
