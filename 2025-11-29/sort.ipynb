```python
x = [9,6,1, 3, 4]
def selection_sort_v1(arr):
    n = len(arr)
    for i in range(n):
        min_index = i
        for j in range(i+1, n):
            if arr[j] < arr[min_index]:
                min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i]
    return arr

print(selection_sort_v1(x))
```

```python
def bubble_sort_v1(arr):
    length = len(arr)        
    for i in range(length):  # 获取轮数
        swapped = False  # 标记本轮是否发生交换
        for j in range(0, length-i-1):  # 每轮比较的次数  9<6
            if arr[j] < arr[j+1]:  # 获取降序排列
                arr[j], arr[j+1] = arr[j+1], arr[j]  # 交换
                swapped = True  # 发生了交换
        if not swapped:  # 如果本轮没有发生交换，说明已经有序，提前结束
            break
    return arr

print(bubble_sort_v1(x))
```

```python
x = [9,6,1, 3, 4]
# 问题：请问该选择排序算法，是不是能象冒泡排序一样，提前结束
def selection_sort_v2(arr):
    length = len(arr)  # 数组长度
    count_iter = 0  # 统计比较次数
    count_swap = 0  # 统计交换次数（选择排序每趟最多一次交换）
    
    for i in range(length-1):
        max_index = i #假设最大数的索引就是当前索引
        
        for j in range(i+1, length):  # j 从 i+1 开始，遍历剩余所有元素
            count_iter += 1  # 每比较一次 → count_iter += 1
            if arr[j] > arr[max_index]:
                max_index = j  # 如果发现新的更大值 → 更新 max_index
                
        if i != max_index:  # 当最大值的索引不等于假设的最大值，进行交换，避免等于也进行交换的次数
            arr[i], arr[max_index] = arr[max_index], arr[i]
            count_swap += 1
    return arr,count_iter, count_swap

print(selection_sort_v2(x))
```
