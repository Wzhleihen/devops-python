```python
import random
import string

product_ids = []
for i in range(1,101):
    new_prat = f'{i:06d}'
    char_part = ''.join(random.choices(string.ascii_lowercase, k=10))
    product_id = f"{new_prat}.{char_part}"
    product_ids.append(product_id)

for pid in product_ids[:10]:
    print(pid)
```
